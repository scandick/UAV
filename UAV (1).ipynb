#Описание объектов симулятора БПЛА

 # **Авионика (Avionics) БПЛА** — это совокупность электронных систем и устройств, используемых в авиации и беспилотных летательных аппаратах (БПЛА) для управления полетом, навигации, связи и других функций. Основные компоненты авионики включают:
   
   **Системы управления полетом (Flight Control Systems)**: Автоматические и полуавтоматические системы, которые помогают управлять самолетом или БПЛА, включая автопилоты и системы стабилизации.
   
   **Навигационные системы (Navigation Systems)**: Включают GPS (Global Positioning System), инерциальные навигационные системы (INS), радионавигационные системы (например, VOR, ILS), которые обеспечивают точное позиционирование и навигацию.
   
   **Системы связи (Communication Systems)**: Радиосистемы для связи с наземными службами и другими летательными аппаратами, включая VHF и UHF радиосвязь, а также системы для передачи данных.
   
   **Системы мониторинга и диагностики (Monitoring and Diagnostic Systems)**: Приборы и системы, которые отслеживают состояние различных систем самолета или БПЛА, включая двигатели, топливные системы и системы управления
   
   **Системы отображения информации (Display Systems)**: Приборные панели и экраны, которые показывают пилоту важную информацию о состоянии самолета, навигации, высоте, скорости и других параметрах.
   
   **Системы предупреждения и аварийные системы (Warning and Emergency Systems)**: Системы, которые предупреждают пилота о потенциальных проблемах, таких как перегрев, низкий уровень топлива, или другие критические состояния.
   
   **Системы управления нагрузкой (Load Management Systems)**: Системы, которые помогают управлять полезной нагрузкой, включая распределение веса и центровку.
   
   **Системы автоматизации (Automation Systems)**: Системы, которые автоматизируют различные аспекты управления полетом и навигации, позволяя снизить нагрузку на пилота.
   
   **Системы обеспечения безопасности (Safety Systems)**: Включают системы резервирования и защиты от отказов, которые обеспечивают безопасность полетов

# 1. **Модель БПЛА**   
   - **Динамика БПЛА**: физика БПЛА, учитывая его массу, инерцию, аэродинамические силы, тягу, сопротивление и управляющие поверхности (mass, inertia, aerodynamic forces, thrust, drag and control surfaces.).
   - **Органы управления**: элероны, рули высоты, рули направления и элементы управления для летательных аппаратов вертикального взлета и посадки (VTOL).
   - **Приборы**: Акселерометры, гироскопы, GPS, альтиметры и камеры. (Accelerometer, gyroscope, GPS, altimeter and camera)
    
    Инерциальный измерительный блок - An Inertial Measure Unit (IMU) — это устройство, способное измерять и сообщать данные об угловом положении (крен, тангаж и рыскание), скорости, изменениях высоты и гравитационных силах, действующих на летательный аппарат.

    An Inertial Measure Unit (IMU) is a device able to measure and report attitude (roll, pitch and yaw), velocity, changes in altitude and gravitational forces acting on an aircraft.


### 2. **Окружающая среда**
   - **Рельеф**: наземные особенности, включая холмы, долины, здания и городские районы, которые могут влиять на динамику полета и навигацию.
   - **Метеоусловия**: переменные, такие как ветер, температура, влажность, осадки и турбулентность, для симуляции реальных условий полета.
   - **Препятствия**: статические и динамические препятствия, такие как деревья, здания, линии электропередач и движущиеся транспортные средства или животные, чтобы усложнить навигацию.

### 3. **Система управления**
   - **Бортовой (авиационный) контроллер**: Реализует алгоритмы управления (например, PID, LQR) для обесечения стабильности и наведения БПЛА.
   - **Автопилот**: Системы, которые обрабатывают автоматизированные задачи полета, такие как навигация по точкам маршрута, удержание высоты и функции возврата на базу.

### 4. **Пользовательский интерфейс**
   - **Элементы управления полетом**: Джойстики, клавиатуры или сенсорные экраны для управления пилотом.
   - **Дисплеи телеметрии**: Визуализируют критически важные данные полета, включая высоту, скорость, заряд батареи и данные с датчиков.
   - **Планировщик миссий**: Инструмент для графической настройки точек маршрута и параметров миссии.

### 5. **Инструменты симуляции**
   - **Физический движок**: Симулирует физику в реальном времени для точного моделирования аэродинамики и механики полета.
   - **Движок рендеринга**: Отображает 3D-визуализации БПЛА, окружающей среды и других объектов в симуляторе.
   - **Журнал событий**: Записывает значимые события, такие как взлеты, посадки и изменения в системе для анализа.

### 6. **Станция управления на земле (GCS Ground Control Station)**
   - **Система связи**: Симулирует передачу данных между БПЛА и наземной станцией управления, включая команды и телеметрические данные.
   - **Пользовательский интерфейс для операторов**: Отображает карты, статус БПЛА и варианты управления для взаимодействия операторов с БПЛА.

### 7. **Типы миссий**
   - **Предопределенные миссии**: Сценарии, такие как поисково-спасательные операции, аэрофотосъемка, обследование или инспекция.
   - **Динамические миссии**: Задачи, которые изменяются в реальном времени в зависимости от условий окружающей среды или вводимых пользователем данных.

### 8. **Сетевые технологии и поддержка нескольких БПЛА**
   - **Координация нескольких БПЛА**: Механизмы для симуляции работы нескольких БПЛА совместно, включая избегание столкновений и полет в формации.
   - **Сетевые технологии**: Системы, позволяющие взаимодействие между несколькими дронами и между дронами и наземной станцией во время симуляции.

### 9. **Инструменты отладки и анализа**
   - **Воспроизведение симуляции**: Возможность воспроизведения симуляции полета для анализа производительности и корректировки параметров контроллера или модели.
   - **Инструменты анализа данных**: Инструменты для анализа данных полета, таких как ускорение, GPS-треки и управляющие команды для повышения точности симуляции.

### 1. PID (Пропорционально-Интегрально-Дифференциальный) контроллер
ПИД-контроллеры — это концепция управления, которая может быть реализована как в программном обеспечении, так и в аппаратном обеспечении. Давайте рассмотрим оба аспекта:

1. Программное обеспечение
Алгоритмы: ПИД-контроллеры часто реализуются в виде алгоритмов в программном обеспечении. Это может быть код, написанный на различных языках программирования (например, C, C++, Python и др.), который выполняет вычисления для управления системой.
Применение: В контексте БПЛА программное обеспечение использует данные от сенсоров (гироскопы, акселерометры, GPS и др.) для вычисления ошибки и управления исполнительными механизмами (моторами, сервоприводами) с помощью ПИД-алгоритма.
2. Аппаратное обеспечение
Контроллеры: В некоторых случаях ПИД-контроллеры могут быть реализованы на уровне аппаратного обеспечения, например, в виде специализированных микроконтроллеров или FPGA (программируемых логических интегральных схем), которые могут выполнять алгоритмы управления.

Интеграция: Многие современные системы управления, включая системы для БПЛА, используют комбинацию аппаратного и программного обеспечения. Например, микроконтроллер может выполнять алгоритмы ПИД-контроллера, которые были загружены в его память.

PID — это один из наиболее распространенных алгоритмов управления, который часто используется в бортовых контроллерах беспилотных летательных аппаратов (БПЛА) для поддержания стабильности и управления полетом.

1. **Роль PID-контроллера**:
   - PID-контроллер используется для регулирования различных параметров полета БПЛА, таких как угол наклона, высота, скорость и курс. Он вычисляет управляющее воздействие (например, команды на двигатели) на основе разности между желаемым состоянием (целевым значением) и текущим состоянием (измеренным значением).
   - PID-контроллер работает по принципу:
     - **Пропорциональная часть (P)**: Реагирует на текущее отклонение от целевого значения.
     - **Интегральная часть (I)**: Учитывает накопленное отклонение за время, чтобы устранить постоянные ошибки.
     - **Дифференциальная часть (D)**: Реагирует на скорость изменения ошибки, что помогает предсказать и сгладить колебания.

2. **Интеграция в бортовом контроллере**:
   - Бортовой контроллер БПЛА включает в себя PID-контроллеры в своих алгоритмах управления для обработки различных режимов полета. Например, для управления углом наклона дронов могут использоваться отдельные PID-контроллеры для управления по осям X, Y и Z.
   - Бортовой контроллер также может использовать другие алгоритмы управления и фильтры, такие как фильтры Калмана, для обработки данных от датчиков и улучшения точности управления.

3. **Настройка и оптимизация**:
   - Эффективность PID-контроллера зависит от правильной настройки коэффициентов P, I и D. В бортовых контроллерах часто используются методы настройки, такие как метод Зиглера-Николса, чтобы добиться оптимальной производительности и стабильности.

4. **Применение в различных режимах полета**:
   - PID-контроллеры могут быть использованы в различных режимах полета, таких как автоматический режим, ручное управление или режимы с различными уровнями автоматизации. Бортовой контроллер управляет переходами между этими режимами и корректирует параметры PID в зависимости от условий полета.



````python
import numpy as np
import matplotlib.pyplot as plt

class PIDController:
    def __init__(self, Kp, Ki, Kd, setpoint, initial_value, num_steps=100, min_output=None, max_output=None):
        self.Kp = Kp  # Proportional gain
        self.Ki = Ki  # Integral gain
        self.Kd = Kd  # Derivative gain
        self.setpoint = setpoint  # Desired setpoint
        self.measured_value = initial_value  # Initial measured value
        self.num_steps = num_steps  # Number of simulation steps

        # PID controller state
        self.integral = 0.0  # Integral term
        self.previous_error = 0.0  # Previous error term

        # Output constraints
        self.min_output = min_output
        self.max_output = max_output

        # Storage for results
        self.control_outputs = []  # Store control outputs
        self.altitude_history = []  # Store measured value history

    def update(self):
        # Calculate the error
        error = self.setpoint - self.measured_value

        # Update integral with windup prevention
        self.integral += error
        if self.min_output is not None and self.integral < self.min_output:
            self.integral = self.min_output
        if self.max_output is not None and self.integral > self.max_output:
            self.integral = self.max_output

        # Calculate derivative
        derivative = error - self.previous_error

        # Calculate control output
        output = self.Kp * error + self.Ki * self.integral + self.Kd * derivative

        # Apply output constraints
        if self.min_output is not None and output < self.min_output:
            output = self.min_output
        if self.max_output is not None and output > self.max_output:
            output = self.max_output

        # Save the previous error for the next iteration
        self.previous_error = error

        # Simulate the dynamics of the system (adjust measured value)
        self.apply_output(output)

        return output

    def apply_output(self, output):
        # Update the measured value based on the control output
        # Here you can add system dynamics simulation if needed
        self.measured_value += output  # For simplicity, this is a direct effect

    def simulate(self):
        # Simulation loop
        for _ in range(self.num_steps):
            output = self.update()  # Update PID controller and get control output
            self.control_outputs.append(output)  # Store output
            self.altitude_history.append(self.measured_value)  # Store measured value history

    def plot_results(self):
        # Plotting control outputs and measured values
        plt.figure(figsize=(12, 6))

        plt.subplot(2, 1, 1)
        plt.plot(self.altitude_history, label='Measured Value', color='b')
        plt.axhline(y=self.setpoint, color='r', linestyle='--', label='Setpoint')
        plt.title('Altitude Response to PID Control')
        plt.xlabel('Time Step')
        plt.ylabel('Altitude')
        plt.legend()
        plt.grid()

        plt.subplot(2, 1, 2)
        plt.plot(self.control_outputs, label='Control Output', color='g')
        plt.title('Control Outputs Over Time')
        plt.xlabel('Time Step')
        plt.ylabel('Control Output')
        plt.legend()
        plt.grid()

        plt.tight_layout()
        plt.show()


# Example usage
setpoint = 10.0  # Desired altitude
initial_value = 0.0  # Initial altitude
Kp, Ki, Kd = 1.0, 0.1, 0.05  # PID constants

# Create the PID controller
pid_controller = PIDController(Kp, Ki, Kd, setpoint, initial_value)

# Run the simulation
pid_controller.simulate()

# Plot the results
pid_controller.plot_results()
````

# 2. Бортовой (авиационный) контроллер:

### Бортовой контроллер (БПЛА) - FlightController
Бортовой контроллер беспилотного летательного аппарата (БПЛА) состоит из нескольких ключевых компонентов, которые работают вместе для управления полетом и обеспечивают выполнение различных задач:

1. **Микроконтроллер или процессор**: Это центральный элемент, который выполняет все вычисления и управление. Он обрабатывает данные от сенсоров и отправляет команды на исполнительные механизмы.

2. **Сенсоры**: Бортовой контроллер использует различные сенсоры для получения информации о состоянии БПЛА:
   - **Гироскопы** и **акселерометры** для определения углового положения и ускорения.
   - **Барометры** для измерения высоты.
   - **GPS** для определения местоположения и скорости.
   - **Магнитометры** для определения направления.

3. **Исполнительные механизмы**: Это устройства, которые приводят в действие элементы управления, такие как сервоприводы для управления рулями или моторами.

4. **Программное обеспечение**: Алгоритмы управления, включая ПИД-контроллеры, которые обрабатывают данные от сенсоров и принимают решения о том, как управлять БПЛА.

5. **Система передачи данных**: Это может быть радиосистема или другие средства связи для передачи данных между БПЛА и наземной станцией, а также для получения команд.

6. **Электропитание**: Бортовой контроллер требует источника питания, который обычно включает аккумуляторы и системы управления энергией.

7. **Дополнительные модули**: В зависимости от назначения БПЛА могут быть установлены дополнительные модули, такие как камеры, датчики расстояния, LiDAR и другие устройства для выполнения специфических задач.

Все эти компоненты работают вместе, чтобы обеспечить стабильный и безопасный полет БПЛА, а также выполнение заданий, таких как съемка, мониторинг или доставка грузов.

````python
import numpy as np
import matplotlib.pyplot as plt

class FlightController:
    def __init__(self, Kp, Ki, Kd, setpoint_x, setpoint_y, setpoint_z, initial_x, initial_y, initial_z, num_steps=100):
        self.pid_x = PIDController(Kp, Ki, Kd, setpoint_x, initial_x, num_steps)
        self.pid_y = PIDController(Kp, Ki, Kd, setpoint_y, initial_y, num_steps)
        self.pid_z = PIDController(Kp, Ki, Kd, setpoint_z, initial_z, num_steps)

    def simulate(self):
        # Run simulation for all axes
        self.pid_x.simulate()
        self.pid_y.simulate()
        self.pid_z.simulate()

    def get_status(self):
        return {
            "X Position": self.pid_x.measured_value,
            "Y Position": self.pid_y.measured_value,
            "Z Position": self.pid_z.measured_value,
            "X Setpoint": self.pid_x.setpoint,
            "Y Setpoint": self.pid_y.setpoint,
            "Z Setpoint": self.pid_z.setpoint,
        }

    def plot_results(self):
        plt.figure(figsize=(12, 8))

        # Subplot for X-axis results
        plt.subplot(3, 1, 1)
        plt.plot(self.pid_x.altitude_history, label='X Position', color='b')
        plt.axhline(y=self.pid_x.setpoint, color='r', linestyle='--', label='X Setpoint')
        plt.title('X Position Response')
        plt.xlabel('Time Step')
        plt.ylabel('Position (X)')
        plt.legend()
        plt.grid()

        # Subplot for Y-axis results
        plt.subplot(3, 1, 2)
        plt.plot(self.pid_y.altitude_history, label='Y Position', color='g')
        plt.axhline(y=self.pid_y.setpoint, color='r', linestyle='--', label='Y Setpoint')
        plt.title('Y Position Response')
        plt.xlabel('Time Step')
        plt.ylabel('Position (Y)')
        plt.legend()
        plt.grid()

        # Subplot for Z-axis results
        plt.subplot(3, 1, 3)
        plt.plot(self.pid_z.altitude_history, label='Z Position', color='c')
        plt.axhline(y=self.pid_z.setpoint, color='r', linestyle='--', label='Z Setpoint')
        plt.title('Z Position Response')
        plt.xlabel('Time Step')
        plt.ylabel('Position (Z)')
        plt.legend()
        plt.grid()

        plt.tight_layout()
        plt.show()

# Example usage
setpoint_x, setpoint_y, setpoint_z = 10.0, 10.0, 10.0  # Desired positions
initial_x, initial_y, initial_z = 0.0, 0.0, 0.0  # Initial positions
Kp, Ki, Kd = 1.0, 0.1, 0.05  # PID constants

# Create the Flight Controller
flight_controller = FlightController(Kp, Ki, Kd, setpoint_x, setpoint_y, setpoint_z, initial_x, initial_y, initial_z, num_steps=100)

# Run the simulation
flight_controller.simulate()

# Plot the results
flight_controller.plot_results()
````

# 2.Симулятор управления БПЛА по X,Y,Z.
**Динамика БПЛА (UAV Dynamics):** Простой класс БПЛА инкапсулирует основные свойства, важные для полета, включая высоту и скорость. Метод обновления интегрирует входные данные по тяге для обновления этих свойств.

**PID контроллер:** Функция `pid_controller` вычисляет выходное управление, используя формулу PID. Она принимает целевое значение (setpoint), измеренное значение (текущая высота) и параметры PID.

**Цикл симуляции:** Основной цикл выполняется фиксированное количество итераций (в данном случае 100). На каждом шаге:

1. PID контроллер вычисляет необходимую тягу для достижения желаемой высоты (setpoint).
2. Состояние БПЛА обновляется с использованием этой тяги.

````python
import numpy as np
import matplotlib.pyplot as plt

class UAV:
    def __init__(self, Kp, Ki, Kd, setpoint_x, setpoint_y, setpoint_z, initial_x, initial_y, initial_z, num_steps=100):
        self.flight_controller = FlightController(Kp, Ki, Kd, setpoint_x, setpoint_y, setpoint_z, initial_x, initial_y, initial_z, num_steps)
        self.simulation_running = False

    def execute_command(self, command):
        parts = command.split()
        if not parts:
            return "ERROR"

        action = parts[0]

        if action == "SET_SETPOINT":
            if len(parts) != 4:
                return "ERROR"
            try:
                x, y, z = map(float, parts[1:4])
                self.flight_controller.pid_x.setpoint = x
                self.flight_controller.pid_y.setpoint = y
                self.flight_controller.pid_z.setpoint = z
                return "SETPOINT_UPDATED"
            except ValueError:
                return "ERROR"

        elif action == "GET_STATUS":
            status = self.flight_controller.get_status()
            return f"STATUS {status['X Position']} {status['Y Position']} {status['Z Position']} {status['X Setpoint']} {status['Y Setpoint']} {status['Z Setpoint']}"

        elif action == "START_SIMULATION":
            self.simulation_running = True
            self.flight_controller.simulate()
            return "SIMULATION_STARTED"

        elif action == "STOP_SIMULATION":
            self.simulation_running = False
            return "SIMULATION_STOPPED"

        elif action == "LOG_DATA":
            return "LOG_DATA <data>"

        elif action == "PING":
            return "PONG"

        else:
            return "ERROR"

    def simulate(self):
        self.flight_controller.simulate()

    def plot_results(self):
        self.flight_controller.plot_results()

    def get_status(self):
        return {
            "X Position": self.flight_controller.pid_x.measured_value,
            "Y Position": self.flight_controller.pid_y.measured_value,
            "Z Position": self.flight_controller.pid_z.measured_value,
            "X Setpoint": self.flight_controller.pid_x.setpoint,
            "Y Setpoint": self.flight_controller.pid_y.setpoint,
            "Z Setpoint": self.flight_controller.pid_z.setpoint,
            "Control Outputs": {
                "X Control": self.flight_controller.pid_x.control_outputs[-1] if self.flight_controller.pid_x.control_outputs else None,
                "Y Control": self.flight_controller.pid_y.control_outputs[-1] if self.flight_controller.pid_y.control_outputs else None,
                "Z Control": self.flight_controller.pid_z.control_outputs[-1] if self.flight_controller.pid_z.control_outputs else None,
            }
        }

# Example usage
setpoint_x, setpoint_y, setpoint_z = 10.0, 10.0, 10.0  # Desired positions
initial_x, initial_y, initial_z = 0.0, 0.0, 0.0  # Initial positions
Kp, Ki, Kd = 1.0, 0.1, 0.05  # PID constants

# Create a UAV instance
uav = UAV(Kp, Ki, Kd, setpoint_x, setpoint_y, setpoint_z, initial_x, initial_y, initial_z, num_steps=100)

# Run the simulation
uav.simulate()

# Plot the results
uav.plot_results()
````

#Имитатор Altimeter Simulator Code

A simple class that simulates altitude readings.
UAV Class: Simulates the UAV that sends altitude data.
GCS Class: Receives altitude data from the UAV.

````python
import socket
import time
import random
import threading
from abc import ABC, abstractmethod

class IAltimeterSimulator(ABC):
    @abstractmethod
    def simulate(self):
        """Simulate altitude readings."""
        pass

class IUAV:
    def send_altitude(self):
        raise NotImplementedError("This method should be overridden.")

class IGCS:
    def receive_altitude(self):
        raise NotImplementedError("This method should be overridden.")

class AltimeterSimulator(IAltimeterSimulator):
    def __init__(self, min_altitude=0, max_altitude=10000):
        self.min_altitude = min_altitude
        self.max_altitude = max_altitude
        self.current_altitude = 0

    def simulate(self):
        """Simulate obtaining a random altitude."""
        while True:
            self.current_altitude = random.uniform(self.min_altitude, self.max_altitude)
            print(f"Simulated altitude: {self.current_altitude:.2f} meters")
            time.sleep(1)  # simulate periodic reporting (every second)

    def get_altitude(self):
        return self.current_altitude

class UAVSimulator(IUAV):
    def __init__(self, host='127.0.0.1', port=14551):
        self.altimeter = AltimeterSimulator()
        self.host = host
        self.port = port
        self.sock = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)

    def send_altitude(self):
        while True:
            altitude = self.altimeter.get_altitude()
            message = f"ALTITUDE:{altitude:.2f}\n"  # Formatting altitude message
            print("+++",message)
            self.sock.sendto(message.encode('utf-8'), (self.host, self.port))
            print(f"Sent: {message.strip()}")
            time.sleep(1)  # Send updated altitude every second

class GCS(IGCS):
    def __init__(self, host='127.0.0.1', port=14551):
        self.host = host
        self.port = port
        self.sock = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
        self.sock.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)  # Allow reuse
        self.sock.bind((self.host, self.port))
        self.running = True  # Flag to control the session

    def receive_altitude(self):
        while self.running:
            try:
                data, _ = self.sock.recvfrom(1024)  # Receive data
                # Process data here (for example, print it)
                print(f"Data received: {data.decode('utf-8')}")
            except socket.error as e:
                print(f"Socket error: {e}")

    def end_session(self):
        self.running = False  # Set the running flag to False
        self.sock.close()  # Close the socket
        print("Session ended. Socket closed.")

# Usage Example
if __name__ == "__main__":
    # Start the altimeter simulation in a thread
    altimeter = AltimeterSimulator()
    altimeter_thread = threading.Thread(target=altimeter.simulate)
    altimeter_thread.start()

    # Initialize the UAV simulator in a thread
    uav = UAVSimulator()
    uav_thread = threading.Thread(target=uav.send_altitude)
    uav_thread.start()

    # Start the Ground Control Station (GCS)
    gcs = GCS()
    gcs_thread = threading.Thread(target=gcs.receive_altitude)
    gcs_thread.start()

    # Simulation of doing something else for 10 seconds
    time.sleep(10)

    # End the session after 10 seconds
    gcs.end_session()
    gcs_thread.join()  # Optionally wait for the thread to finish

    # Optionally join UAV thread if needed
    # uav_thread.join()

    print("GCS thread has exited.")
````

# Обзор протокола MAVLink между БПЛА и наземной станцией управления (GCS)

Протокол связи между БПЛА (беспилотным летательным аппаратом) и наземной станцией управления (GCS) играет ключевую роль в обеспечении эффективного контроля, телеметрии и обмена данными. Один из наиболее широко используемых протоколов для этой цели — это протокол MAVLink (Micro Air Vehicle Link). Ниже представлен обзор протокола и пример его реализации.

### Обзор протокола MAVLink
1. **Структура сообщения**: Сообщения MAVLink состоят из заголовка, полезной нагрузки и контрольной суммы. Заголовок включает информацию, такую как длина сообщения, номер последовательности, ID системы и ID компонента.

2. **Типы сообщений**: MAVLink поддерживает множество типов сообщений, включая:
   - **Heartbeat (Сердцебиение)**: Указывает, что БПЛА активен и предоставляет информацию об его статусе.
   - **Данные GPS**: Передает информацию об местоположении БПЛА.
   - **Данные о положении**: Предоставляет информацию о ориентации БПЛА.
   - **Командные сообщения**: Используются для отправки команд БПЛА (например, взлет, посадка).

3. **Каналы связи**: MAVLink может работать через различные каналы связи, включая последовательные порты, UDP и TCP.

4. **Режимы**: БПЛА может работать в различных режимах (например, ручной, автоматический, управляемый), и GCS может изменять эти режимы с помощью специальных команд.

# Формат сообщений:
Сообщение состоит из следующей структуры:
[STX] [Payload Length] [Sequence] [System ID] [Component ID] [Message ID] [Payload] [Checksum]
STX: Стартовый байт (0xFE).
Payload Length: Длина полезной нагрузки.
Sequence: Номер последовательности сообщения.
System ID: Идентификатор системы (БПЛА).
Component ID: Идентификатор компонента (например, контроллер).
Message ID: Идентификатор типа сообщения.
Payload: Данные сообщения.
Checksum: Контрольная сумма для проверки целостности данных.

### Примечание
Каждая команда может иметь свои параметры и может быть использована в различных контекстах. Полный список команд и их описание можно найти в официальной документации MAVLink, которая доступна на сайте [MAVLink Documentation](https://mavlink.io/en/).


Команды MAVLink задаются через отправку сообщений, которые содержат соответствующие данные для выполнения определенного действия. Каждое сообщение имеет свою структуру и формат. Давайте рассмотрим, как можно отправить команды через протокол MAVLink с использованием примеров.

# Другие протоколы
Cуществует несколько других протоколов, которые используются для связи между беспилотными летательными аппаратами (БПЛА) и наземными станциями управления (GCS):

### 1. **TELEMETRY (Telem)**
Протокол телеметрии используется для передачи данных о состоянии БПЛА и его местоположении. Он может включать в себя различные типы сообщений, такие как данные о высоте, скорости, состоянии батареи и т.д. Этот протокол может быть как простым, так и сложным в зависимости от реализации.

### 2. **FrSky**
FrSky — это протокол, который часто используется в радиоуправляемых моделях. Он позволяет передавать данные о состоянии БПЛА на пульт управления и обратно. FrSky поддерживает передачу данных о телеметрии, что делает его популярным среди любителей моделирования.

### 3. **SBUS**
SBUS — это протокол, используемый для передачи сигналов управления от передатчика к приемнику. Он поддерживает многоканальную передачу и может использоваться для управления несколькими сервоприводами и другими устройствами.

### 4. **DShot**
DShot — это цифровой протокол, используемый для передачи команд управления к бесколлекторным моторам. Он обеспечивает более высокую точность и надежность по сравнению с аналоговыми протоколами, такими как PWM.

### 5. **ROS (Robot Operating System)**
ROS — это не протокол в традиционном смысле, а фреймворк для разработки программного обеспечения для робототехники. Он поддерживает обмен сообщениями между различными компонентами системы, включая БПЛА и GCS. ROS может использовать различные транспортные протоколы, такие как TCP/IP и UDP.

### 6. **MAVLink 2.0**
Хотя это и расширение MAVLink, стоит упомянуть, что MAVLink 2.0 добавляет новые функции по сравнению с первой версией, такие как поддержка расширяемых сообщений и улучшенная безопасность.

### 7. **CAN (Controller Area Network)**
CAN используется в некоторых системах для обмена данными между различными компонентами БПЛА, такими как контроллеры двигателей и сенсоры. Он обеспечивает высокую надежность и устойчивость к помехам.

### 8. **UDP/TCP**
Хотя это не протокол для БПЛА, использование стандартных сетевых протоколов, таких как UDP и TCP, также может быть применимо для передачи данных между БПЛА и GCS, особенно в случае, если используется интернет-соединение или локальная сеть.

### Вывод
Выбор протокола зависит от конкретных требований проекта, таких как дальность связи, скорость передачи данных, надежность и доступные ресурсы. Каждый из протоколов имеет свои преимущества и недостатки, поэтому важно учитывать их при разработке системы управления БПЛА.

````python
import socket
import json
import time
import random

# Настройки соединения
host = '192.168.1.100'  # IP адрес БПЛА или наземной станции
port = 14550  # Порт, на который отправляются данные
sock = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)

# Время работы имитатора в секундах
simulation_duration = 10  # Установите желаемую продолжительность имитации
start_time = time.time()  # Запоминаем время начала имитации

def simulate_sensor_data():
    """Функция для имитации данных о сенсорах"""
    latitude = 51.5074 + random.uniform(-0.01, 0.01)  # Имитируем небольшие изменения
    longitude = 0.1278 + random.uniform(-0.01, 0.01)
    altitude = 500 + random.randint(-10, 10)  # Небольшие колебания высоты
    status = 'OK' if random.random() > 0.1 else 'ERROR'  # Имитация статуса
    return {
        'latitude': latitude,
        'longitude': longitude,
        'altitude': altitude,
        'status': status
    }

if __name__ == "__main__":
    while True:
        current_time = time.time()  # Получаем текущее время
        elapsed_time = current_time - start_time  # Вычисляем прошедшее время

        if elapsed_time > simulation_duration:
            print("Simulation time is over.")
            break  # Прекращаем выполнение, если время имитации истекло

        # Получаем имитированные данные
        data = simulate_sensor_data()

        # Отправляем данные в формате JSON
        try:
            sock.sendto(json.dumps(data).encode(), (host, port))
            print(f"Sent data: {data}")
        except Exception as e:
            print(f"Error sending data: {e}")

        # Задержка перед следующей отправкой данных
        time.sleep(1)  # Отправляем данные каждую секунду
````

Реализаци с json

````python
import socket
import time
import random
import threading
import json
from abc import ABC, abstractmethod

class IAltimeterSimulator(ABC):
    @abstractmethod
    def simulate(self):
        pass

class IUAV:
    def send_altitude(self):
        raise NotImplementedError("This method should be overridden.")

class IGCS:
    def receive_altitude(self):
        raise NotImplementedError("This method should be overridden.")

class AltimeterSimulator(IAltimeterSimulator):
    def __init__(self, min_altitude=0, max_altitude=10000):
        self.min_altitude = min_altitude
        self.max_altitude = max_altitude
        self.current_altitude = 0

    def simulate(self):
        while True:
            self.current_altitude = random.uniform(self.min_altitude, self.max_altitude)
            print(f"Simulated altitude: {self.current_altitude:.2f} meters")
            time.sleep(1)  # simulate periodic reporting (every second)

    def get_altitude(self):
        return self.current_altitude

class UAVSimulator(IUAV):
    def __init__(self, host='127.0.0.1', port=14551):
        self.altimeter = AltimeterSimulator()
        self.host = host
        self.port = port
        self.sock = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)

    def send_altitude(self, duration):
        end_time = time.time() + duration
        while time.time() < end_time:
            altitude = self.altimeter.get_altitude()
            # Ensure the altitude value is valid
            print(f"Current altitude for sending: {altitude:.2f} meters")
            message = json.dumps({"altitude": altitude})
            self.sock.sendto(message.encode('utf-8'), (self.host, self.port))
            print(f"Sent: {message.strip()}")
            time.sleep(1)  # Send updated altitude every second

class GCS(IGCS):
    def __init__(self, host='127.0.0.1', port=14551):
        self.host = host
        self.port = port
        self.sock = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
        self.sock.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)
        self.sock.bind((self.host, self.port))
        self.running = True

    def receive_altitude(self):
        while self.running:
            try:
                data, _ = self.sock.recvfrom(1024)
                if data:
                    # Decode JSON and process data here
                    altitude_data = json.loads(data.decode('utf-8'))
                    print(f"Data received: {altitude_data}")
                else:
                    print("No data received.")
            except socket.error as e:
                print(f"Socket error: {e}")
            except json.JSONDecodeError as e:
                print(f"JSON decode error: {e}: Data was: {data.decode('utf-8')}")
                print(f"Raw data received (bytes): {data}")  # Add this for debugging

    def end_session(self):
        self.running = False
        self.sock.close()
        print("Session ended. Socket closed.")

# Main execution
if __name__ == "__main__":
    # Start the altimeter simulation in a thread
    altimeter = AltimeterSimulator()
    altimeter_thread = threading.Thread(target=altimeter.simulate)
    altimeter_thread.start()

    # Initialize the UAV simulator in a thread
    uav = UAVSimulator()
    simulation_duration = 10  # Time for sending altitudes in seconds
    uav_thread = threading.Thread(target=uav.send_altitude, args=(simulation_duration,))
    uav_thread.start()

    # Start the Ground Control Station (GCS)
    gcs = GCS()
    gcs_thread = threading.Thread(target=gcs.receive_altitude)
    gcs_thread.start()

    # Wait for the UAV simulator to finish
    uav_thread.join()

    # End the GCS session after UAV simulation is complete
    gcs.end_session()
    gcs_thread.join()

    print("GCS thread has exited.")
````

# 1.Модель БПЛА с двигателями и без них.

### Модели БПЛА без двигателей:
1. **Модели на основе аэростатов**:
   - **Тепловые аэростаты**: Используют горячий воздух для подъема.
   - **Гелиевые шары**: Заполнены гелием, поднимаются благодаря легкости.
2. **Парапланы и дельтапланы**: Эти аппараты полагаются на подъемную силу, создаваемую за счет воздушных потоков и не имеют двигателей.
3. **Системы дронов с использованием воздушных подушек**: Построены так, чтобы использовать компрессионные технологии для подъема в воздух.
4. **Скользящие дроны**: Модели, которые запускаются с высоты и могут скользить на большие расстояния, используя аэродинамику.

### Модели БПЛА с двигателями:
1. **Квадрокоптеры**: Примеры: DJI Phantom 4, DJI Mavic Air 2. Эти дроны имеют несколько моторов и широко используются для фотосъемки и видеосъемки.
2. **МультироTorы**: Примеры: Yuneec Typhoon H, Parrot Anafi. Эти дроны также используют несколько пропеллеров для подъема и маневрирования.
3. **Самолетные дроны**: Примеры: DJI Matrice 300 RTK, SenseFly eBee X. Эти модели обычно используются в сельском хозяйстве и геодезии и имеют фиксированное крыло.
4. **Гибридные дроны**: Например, дроны, способные взлетать и садиться вертикально (VTOL), как Quantum Systems Trinity F90+.
5. **Солнечные дроны**: Пример: Airbus Zephyr. Эти БПЛА используют солнечные панели для получения энергии и могут летать на больших высотах в течение длительного времени.

### Виды роторных систем в БПЛА:
1. **Один ротор**: Используется в вертолетах и некоторых БПЛА. Система с одним ротором состоит из одного главного ротора для подъема и хвостового ротора для стабильности и управления.
2. **Много роторные**: Включает квадрокоптеры (четыре ротора), гексакоптеры (шесть роторов) и окто-коптеры (восемь роторов). Эти конструкции обеспечивают повышенную стабильность, резервирование и маневренность. Они обычно используются для аэрофотосъемки, наблюдения и развлекательных полетов.
3. **Коаксиальные роторы**: Эта конструкция включает два ротора, установленные на одной оси, вращающиеся в противоположных направлениях. Коаксиальные роторы часто встречаются в современных конструкциях дронов и увеличивают подъем и эффективность.

### Функции роторов:
- **Генерация подъемной силы**: Роторы создают подъемную силу, вращаясь и создавая разницу в давлении воздуха над и под лопастями.
- **Привод**: Изменяя скорость и угол поворота роторов, БПЛА может двигаться вперед, назад и вбок.
- **Управление**: Позиция и скорость роторов помогают поддерживать стабильность и обеспечивают точное маневрирование.

«ротор» конкретно относится к вращающимся лопаткам, которые обеспечивают подъем и управление для БПЛА, они действительно являются критически важным компонентом многих конструкций БПЛА и могут считаться частью системы «двигателя», особенно в много роторных дронах.


Наибольшее распространение среди беспилотных летательных аппаратов (БПЛА) имеют мультикоптеры, особенно квадрокоптеры. Вот несколько основных типов БПЛА, которые широко используются:

### 1. Квадрокоптеры
Имеют четыре ротора, расположенные в форме квадрата. Это самый популярный тип БПЛА, используемый как в коммерческих, так и в любительских целях.
- **Преимущества**: Простота управления, стабильность при полетах, возможность выполнять вертикальный взлет и посадку (VTOL).
- **Применение**: Фотосъемка, видеосъемка, мониторинг сельского хозяйства, доставка товаров и многое другое.

### 2. Гексакоптеры
Имеют шесть роторов. Обеспечивают большую грузоподъемность и надежность по сравнению с квадрокоптерами.
- **Преимущества**: Более высокая устойчивость и возможность продолжать полет даже при выходе из строя одного из роторов.
- **Применение**: Профессиональная аэросъемка, спасательные операции и работы, требующие больших нагрузок.

### 3. Октакаптеры
Имеют восемь роторов. Это тип, который чаще всего используется для профессиональных и специализированных задач.
- **Преимущества**: Возможность не только нести большие нагрузки, но и функционировать в случае выхода из строя нескольких моторов.
- **Применение**: Профессиональная фотосъемка, карты для 3D-моделирования и другие технические задания.

### 4. Винтокрылые БПЛА
Комбинация фиксированного крыла и вращающихся лопастей. Это позволяет им комбинировать преимущества самолетов и вертолетов.
- **Преимущества**: Большая дальность полета и более высокая скорость по сравнению с мультикоптерами.
- **Применение**: Долгосрочные наблюдения, сельское хозяйство, картография.

### 5. Фиксированные крыла. Традиционная конструкция самолета с фиксированным крылом.
- **Преимущества**: Эффективность при длительных полетах и больших расстояниях, меньшая потребление энергии по сравнению с мультикоптерами.
- **Применение**: Геодезия, мониторинг окружающей среды, агрономические исследования.

# Тип роторной системы:
 1. Фиксированные роторы: Лопасти остаются в фиксированном положении и не изменяют угол атаки во время полета. Применяются в простых моделях и некоторых типах вертолетов.

2. Роторы с изменяемым углом атаки: Лопасти могут изменять угол атаки, что позволяет управлять подъемной силой и маневренностью. Используются в большинстве современных вертолетов.

3. Многоплечевые роторы: Системы с несколькими лопастями (например, 3, 4 или более). Обеспечивают большую подъемную силу и стабильность. Применяются в многоплечевых дронов и вертолетах.

4. Сдвоенные роторы: Используются в некоторых вертолетах, где два ротора вращаются в противоположных направлениях для компенсации крутящего момента.

# По функциональности
1. Основные роторы: Основные роторные системы, отвечающие за подъемную силу и маневры (например, роторы вертолетов).

2. Тандемные роторы: Два ротора, расположенные один над другим или один за другим, которые работают совместно для создания подъемной силы.

3. Косые роторы: Используются в некоторых летательных аппаратах для управления направлением движения и подъемной силой.

### Особенности фиксированных роторных систем квадрокоптеров:
1. **Постоянный угол атаки**: Лопасти роторов квадрокоптера имеют фиксированный угол наклона, что упрощает конструкцию и управление.
2. **Управление подъемной силой**: путем изменения скорости вращения каждого из роторов. Увеличение скорости одного или нескольких роторов приводит к подъему, в то время как уменьшение скорости вызывает снижение.
3. **Симметричное расположение**: В квадрокоптерах обычно четыре ротора расположены симметрично (в форме квадрата или ромба), что обеспечивает стабильность и управляемость.
4. **Электронные системы управления**: Квадрокоптеры используют электронные системы управления для регулирования скорости вращения роторов, что позволяет им выполнять различные маневры, такие как повороты, наклоны и вертикальный подъем/спуск.

Асинхронная симуляция: запуск каждого ротора в отдельных потоках, чтобы симуляция роторов выполнялась параллельно.

Метода для изменения подъемной силы: Чтобы можно было в динамике изменять скорость роторов и наблюдать за изменениями подъемной силы.

Визуализация данных (в простом текстовом виде): Для удобства можно и далее выводить информацию о статусе, но также добавим возможность изменять скорости роторов в рамках симуляции, чтобы видеть изменения подъемной силы.

````python
import time
import threading

class Rotor:
    def __init__(self, id):
        self.id = id
        self.speed = 0  # Скорость вращения ротора в об/мин
        self.lift = 0  # Подъемная сила

    def set_speed(self, speed):
        """Устанавливает скорость вращения ротора."""
        self.speed = speed
        print(f"Ротор {self.id}: скорость установлена на {self.speed} об/мин")

    def get_lift(self):
        """Возвращает подъемную силу, создаваемую ротором."""
        # Для простоты предположим, что подъемная сила пропорциональна скорости
        lift = self.speed * 0.1  # Коэффициент для преобразования скорости в подъемную силу
        self.lift = lift
        return lift

    def simulate(self, duration, interval=1):
        """Имитация работы ротора в течение заданного времени."""
        print(f"Начинаем симуляцию работы ротора {self.id} на скорости {self.speed} об/мин на {duration} секунд.")
        for t in range(0, duration, interval):
            lift = self.get_lift()
            print(f"Время: {t} сек - Ротор {self.id}: скорость {self.speed} об/мин, подъемная сила {lift:.2f} Н")
            time.sleep(interval)  # Пауза для имитации времени


class Quadcopter:
    def __init__(self):
        # Создаем 4 ротора
        self.rotors = [Rotor(i) for i in range(4)]
        self.lock = threading.Lock()  # Для безопасной работы с несколькими потоками

    def set_rotor_speeds(self, speeds):
        """Устанавливает скорости для всех роторов."""
        for i, speed in enumerate(speeds):
           # print("i = ",i)
           # print("speed = ",speed)
            self.rotors[i].set_speed(speed)

    def calculate_total_lift(self):
        """Вычисляет общую подъемную силу от всех роторов."""
        total_lift = sum(rotor.get_lift() for rotor in self.rotors)
        return total_lift

    def display_status(self):
        """Выводит статус квадрокоптера и подъемную силу."""
        print("Статус квадрокоптера:")
        for rotor in self.rotors:
            print(f"Ротор {rotor.id}: скорость {rotor.speed} об/мин, подъемная сила {rotor.get_lift():.2f} Н")
        print(f"Общая подъемная сила: {self.calculate_total_lift():.2f} Н")

    def start_simulation(self, duration):
        """Запускает симуляцию работы всех роторов в отдельных потоках."""
        threads = []
        for rotor in self.rotors:
            thread = threading.Thread(target=rotor.simulate, args=(duration,))
            threads.append(thread)
            thread.start()

        for thread in threads:
            thread.join()  # Ожидаем завершения всех потоков


# Пример использования
if __name__ == "__main__":
    quadcopter = Quadcopter()

    # Устанавливаем скорость для всех роторов
    quadcopter.set_rotor_speeds([4000, 4000, 4000, 4000])
    quadcopter.display_status()  # Выводим статус квадрокоптера

    # Запускаем симуляцию для каждого ротора
    quadcopter.start_simulation(duration=5)  # Симуляция работы всех роторов в течение 5 секунд

    # Изменение скорости роторов для тестирования динамики
    print("\nИзменяем скорости роторов.")
    quadcopter.set_rotor_speeds([5000, 5000, 3000, 3000])  # Устанавливаем разные скорости для маневра
    quadcopter.display_status()  # Отображение нового статуса
    quadcopter.start_simulation(duration=5)  # Запускаем новую симуляцию
````

# Управление роторами квадрокоптера
1. **Контроллер полета** - центральный элемент управления БПЛА. Он отвечает за обработку данных от различных сенсоров (гироскопы, акселерометры, GPS и т.д.) и принимает решения о том, как регулировать скорость вращения каждого ротора для достижения желаемого поведения (подъем, спуск, поворот и т.д.). Контроллер полета использует алгоритмы управления, такие как PID-контроль, для поддержания стабильности и управления движением.
2. **Электронные регуляторы скорости (ESC)**: управляет каждым ротором, получая команды от контроллера полета и регулируя мощность, подаваемую на моторы.
   - Они преобразуют сигналы из контроллера полета в PWM (широтно-импульсная модуляция) для управления скоростью вращения моторов.
3. **Сенсоры**: информацию о текущем состоянии аппарата. Например, гироскопы и акселерометры измеряют угловые скорости и ускорения, а GPS предоставляет информацию о местоположении. Эти данные используются для определения положения и ориентации БПЛА в пространстве, что критично для управления его движением.
4. **Пульт управления и наземная станция**:  Оператор может управлять БПЛА с помощью пульта дистанционного управления или наземной станции, которые отправляют команды на контроллер полета. Это позволяет оператору изменять параметры полета, такие как высота, направление и скорость.


#Имитации углов рыскания (yaw), тангажа (pitch) и крена (roll) квадрокоптера.

 Углы отражают ориентацию квадрокоптера в пространстве и управляют его маневрами.

Основные концепции
Угол рыскания (Yaw): Поворот вокруг вертикальной оси. Управляется изменением скорости отдельных роторов.
Угол тангажа (Pitch): Поворот вокруг горизонтальной оси, которая проходит от одной стороны квадрокоптера до другой. Угол тангажа можно изменять, меняя скорость передних и задних роторов.

Угол крена (Roll): Поворот вокруг оси, которая проходит через переднюю и заднюю части квадрокоптера. Угол крена изменяется за счет изменения скорости левых и правых роторов.

Класс Rotor: Остается без изменений, так как он отвечает за управление скоростью и подъемной силой роторов.

Класс Quadcopter:
Три новых свойства: yaw, pitch, roll, которые представляют углы рыскания, тангажа и крена.
Метод update_orientation позволяет изменять углы в зависимости от заданных изменений.
Метод simulate_flight иллюстрирует периодическое обновление угла тангажа.

Управление ориентацией
В реальном приложении углы могут изменяться в зависимости от введенных команд с пульта управления или для выполнения определенных маневров. Например, увеличивая скорость передних роторов и уменьшая скорость задних, можно тут же изменить угол тангажа.

Для более точной симуляции углов рыскания, тангажа и крена можно также использовать модели, учитывающие физику и аэродинамику, но это требует гораздо более сложной реализации.


````python
import time
import threading

class Rotor:
    def __init__(self, id):
        self.id = id
        self.speed = 0  # Скорость вращения ротора в об/мин

    def set_speed(self, speed):
        """Устанавливает скорость вращения ротора."""
        self.speed = speed
        print(f"Ротор {self.id}: скорость установлена на {self.speed} об/мин")

    def get_lift(self):
        """Возвращает подъемную силу, создаваемую ротором."""
        return self.speed * 0.1  # Коэффициент для преобразования скорости в подъемную силу


class Quadcopter:
    def __init__(self):
        self.rotors = [Rotor(i) for i in range(4)]
        self.yaw = 0  # Угол рыскания
        self.pitch = 0  # Угол тангажа
        self.roll = 0  # Угол крена

    def set_rotor_speeds(self, speeds):
        """Устанавливает скорости для всех роторов."""
        for i, speed in enumerate(speeds):
            self.rotors[i].set_speed(speed)

    def calculate_total_lift(self):
        """Вычисляет общую подъемную силу от всех роторов."""
        return sum(rotor.get_lift() for rotor in self.rotors)

    def update_orientation(self, yaw_change, pitch_change, roll_change):
        """Изменяет углы рыскания, тангажа и крена."""
        self.yaw += yaw_change
        self.pitch += pitch_change
        self.roll += roll_change

        # Простой вывод текущих углов
        print(f"Текущая ориентация - Тангаж: {self.pitch:.2f}°, Крен: {self.roll:.2f}°, Рыскание: {self.yaw:.2f}°")

    def display_status(self):
        """Выводит статус квадрокоптера и подъемную силу."""
        print("Статус квадрокоптера:")
        for rotor in self.rotors:
            print(f"Ротор {rotor.id}: скорость {rotor.speed} об/мин, подъемная сила {rotor.get_lift():.2f} Н")
        print(f"Общая подъемная сила: {self.calculate_total_lift():.2f} Н")

    def simulate_flight(self, duration):
        """Имитация полета квадрокоптера."""
        start_time = time.time()
        while time.time() - start_time < duration:
            # Пример изменения углов в зависимости от скорости роторов
            if self.rotors[0].speed > self.rotors[1].speed:
                self.update_orientation(0, 1, 0)  # Увеличиваем тангаж при увеличении скорости переднего ротора
            if self.rotors[0].speed < self.rotors[1].speed:
                self.update_orientation(0, -1, 0)  # Уменьшаем тангаж, если передний ротор медленнее

            if self.rotors[2].speed > self.rotors[3].speed:
                self.update_orientation(0, 0, 1)  # Увеличиваем крен при увеличении скорости левого ротора
            if self.rotors[2].speed < self.rotors[3].speed:
                self.update_orientation(0, 0, -1)  # Уменьшаем крен, если левый ротор медленнее

            # Для простоты добавим небольшой интервал времени перед очередным обновлением
            time.sleep(1)  # Имитация интервала времени


# Пример использования
if __name__ == "__main__":
    quadcopter = Quadcopter()

    # Устанавливаем скорость для всех роторов
    quadcopter.set_rotor_speeds([4000, 4000, 4000, 4000])
    quadcopter.display_status()  # Выводим статус квадрокоптера

    # Запускаем симуляцию полета
    quadcopter.simulate_flight(duration=10)  # Симуляция полета в течение 10 секунд

    # Изменение скорости роторов для следующей симуляции
    print("\nИзменяем скорости роторов для маневра.")
    quadcopter.set_rotor_speeds([5000, 4000, 3000, 4000])  # Пример маневра
    quadcopter.display_status()  # Отображение нового статуса

    # Запускаем новую симуляцию полета
    quadcopter.simulate_flight(duration=10)  # Симуляция полета в течение 10 секунд
````

#Модели квадроторора и управление квадророторами.

###Различают линейные и не линейные модели.
## 1. Линейная модель динамика системы аппроксимируется линейными уравнениями.
В линейных моделях предполагается, что изменения в выходных переменных (например, углы крена, тангажа и высота) пропорциональны изменениям в управляющих воздействиях (например, подъемной силе и моментах от роторов). Это позволяет использовать более простые математические методы и алгоритмы управления, такие как PID-контроллеры.

### Основные отличия между линейной и нелинейной моделями:
1.  **Поведение**
 **Линейность**: система ведет себя линейно в малых диапазонах изменений. Например, если вы измените подъемную силу на определенное значение, то изменение высоты будет пропорционально этому изменению. Линейные уравнения имеют свойства суперпозиции (сумма решений является решением).
**Нелинейная модель**:  система может вести себя нелинейно, особенно при больших изменениях. Например, подъемная сила может не пропорционально влиять на высоту при больших углах наклона или скоростях.
2. **Сложность**:
**Линейная модель**: проще в анализе и управлении, так как линейные уравнения легче решать и анализировать. Это позволяет использовать простые методы управления.
**Нелинейная модель**: сложнее в анализе и управлении, так как требует учета нелинейных взаимодействий, что требует более сложных методов управления, таких как адаптивное или нелинейное управление.
3. **Область применения**:
**Линейная модель**: Обычно используется для малых отклонений от равновесия. Например, для небольших изменений углов крена и тангажа.
**Нелинейная модель**: Применяется, когда система работает в широком диапазоне состояний или когда требуется учитывать сложные динамические эффекты, такие как взаимодействие между роторами, влияние внешних условий и т.д.
4. **Прогнозирование**:
**Линейная модель**: эффективне для прогнозирования поведения системы в ограниченных диапазонах, но может давать неверные результаты при больших изменениях.
**Нелинейная модель**: точнее для прогнозирования поведения системы в широком диапазоне состояний, но требует больше вычислительных ресурсов.

Вот основные формулы для линейной и нелинейной модели квадроторора:

# Линейная модель
Линейная модель квадроторора обычно описывается с помощью линейных дифференциальных уравнений, которые могут быть получены из уравнений Ньютона-Эйлера. Основные уравнения движения квадроторора можно записать в виде:
1. **Уравнение движения по вертикали (высота)**:
   $$
   m \ddot{z} = T - mg
   $$
   где:
   - $m$ — масса квадроторора,
   - $\ddot{z}$ — вертикальное ускорение,
   - $T$ — подъемная сила,
   - $g$ — ускорение свободного падения.
2. **Уравнения углового движения**:
   Используя моменты, действующие на квадротор, можно описать динамику вращения:
   $$
   I \dot{\omega} = \tau
   $$
   где:
   - $I$ — момент инерции,
   - $\dot{\omega}$ — угловое ускорение,
   - $\tau$ — момент, создаваемый роторами.
3. **PID-контроллер для управления**:
   $$
   u = -K_p e(t) - K_d \dot{e}(t) - K_i \int_0^t{e(\tau)}d\tau
   $$
   где:
   - $u$ — управляющее воздействие,
   - $K_p, K_i, K_d$ — коэффициенты для пропорциональной, интегральной и дифференциальной частей,
   - $e(t)$ — ошибка системы.

# Вывод
Линейные модели проще в анализе и управлении, но ограничены в точности при больших отклонениях. Нелинейные модели более сложны, но обеспечивают более точное описание динамики системы в широком диапазоне условий. Выбор модели зависит от конкретной задачи и требований к точности.


````python
import matplotlib.pyplot as plt
class QuadcopterLinear:
    def __init__(self, mass, inertia):
        self.mass = mass
        self.inertia = inertia

        # Начальное состояние
        self.z = 0.0  # Высота
        self.z_dot = 0.0  # Вертикальная скорость

        # PID параметры
        self.Kp = 2.0  # Увеличено для большей подъемной силы
        self.Ki = 0.1
        self.Kd = 0.5
        self.error_integral = 0.0
        self.previous_error = 0.0
    def compute_control_input(self, target_z):
        """Вычисляет управляющее воздействие с помощью PID-контроллера"""
        error = target_z - self.z
        self.error_integral += error
        error_derivative = error - self.previous_error

        # Управляющее воздействие
        u = self.Kp * error + self.Kd * error_derivative + self.Ki * self.error_integral
        self.previous_error = error

        # Ограничение на максимальную подъемную силу
        max_thrust = 20.0  # Максимальная подъемная сила
        if u > max_thrust:
            u = max_thrust
        return u
    def update_state(self, control_input, dt):
        """Обновляет состояние квадрокоптера с учетом управления"""
        g = 9.81  # Ускорение свободного падения
        T = control_input  # Подъемная сила

        # Уравнение движения по вертикали
        acceleration = (T - self.mass * g) / self.mass  # Вертикальное ускорение
        self.z_dot += acceleration * dt  # Обновление вертикальной скорости
        self.z += self.z_dot * dt  # Обновление высоты

        # Проверка на отрицательную высоту
        if self.z < 0:
            self.z = 0
            self.z_dot = 0  # Остановить движение вниз при достижении земли
    def simulate(self, target_z, total_time):
        """Симуляция движения квадрокоптера"""
        dt = 0.01  # Уменьшен шаг времени для большей точности
        time_steps = int(total_time / dt)
        z_history = []

        for _ in range(time_steps):
            control_input = self.compute_control_input(target_z)
            self.update_state(control_input, dt)
            z_history.append(self.z)
#            print(f"Высота: {self.z:.2f}, Вертикальная скорость: {self.z_dot:.2f}, Управление: {control_input:.2f}")  # Логирование данных

        return z_history
# Пример использования линейной модели
if __name__ == "__main__":
    quadcopter = QuadcopterLinear(mass=1.0, inertia=0.1)  # Масса 1 кг, момент инерции 0.1
    target_height = 5.0  # Желаемая высота
    height_history = quadcopter.simulate(target_z=target_height, total_time=10)  # Симуляция в течение 10 секунд

    # Визуализация результатов
    plt.plot(height_history, label='Высота (z)')
    plt.axhline(y=target_height, color='r', linestyle='--', label='Целевая высота')
    plt.title("Симуляция линейной модели квадрокоптера")
    plt.xlabel("Шаг времени")
    plt.ylabel("Высота (м)")
    plt.legend()
    plt.show()
````

# Нелинейная модель
Нелинейная модель квадроторора учитывает более сложные взаимодействия и может быть описана с помощью нелинейных уравнений. Основные уравнения могут выглядеть следующим образом:
1. **Уравнение движения по вертикали (высота)**:
   $$
   m \ddot{z} = T - mg - f(z, \dot{z})
   $$
   где:
   - $f(z, \dot{z})$ — нелинейные силы, такие как сопротивление воздуха, которые зависят от высоты $z$ и скорости $\dot{z}$.
2. **Уравнения углового движения**:
   Нелинейные уравнения могут включать взаимодействия между углами:
   $$
   I \dot{\omega} = \tau + g(\phi, \theta, \psi)
   $$
   где:
   - $g(\phi, \theta, \psi)$ — нелинейные функции, зависящие от угловых координат.
3. **Уравнение состояния системы**:
   Нелинейные модели могут быть описаны с помощью уравнения состояния:
   $$
   \dot{x} = f(x, u)
   $$
   где:
   - $x$ — вектор состояния (например, углы и высота),
   - $u$ — вектор управляющих воздействий,
   - $f(x, u)$ — нелинейная функция, описывающая динамику системы.


````python
import numpy as np
import matplotlib.pyplot as plt
class QuadcopterNonlinear:
    def __init__(self, mass, inertia):
        self.mass = mass
        self.inertia = inertia

        # Начальное состояние
        self.z = 0.0  # Высота
        self.z_dot = 0.0  # Вертикальная скорость

        # PID параметры
        self.Kp = 3.0
        self.Ki = 0.1
        self.Kd = 0.5
        self.error_integral = 0.0
        self.previous_error = 0.0

        # Параметры ветра
        self.wind_speed = 0.0  # Скорость ветра
        self.wind_variation = 0.5  # Степень изменения ветра (м/с)

    def compute_control_input(self, target_z):
        """Вычисляет управляющее воздействие с помощью PID-контроллера"""
        error = target_z - self.z
        self.error_integral += error
        error_derivative = error - self.previous_error

        # Управляющее воздействие
        u = self.Kp * error + self.Kd * error_derivative + self.Ki * self.error_integral
        self.previous_error = error

        # Нелинейный эффект: добавление сопротивления
        drag = 0.1 * self.z_dot ** 2  # Простая модель сопротивления
        u = max(0, u - drag)  # Убедитесь, что управление не отрицательное

        return u
    def update_wind(self):
        """Обновляет скорость ветра"""
        self.wind_speed += np.random.uniform(-self.wind_variation, self.wind_variation)

    def update_state(self, control_input, dt):
        """Обновляет состояние квадрокоптера с учетом управления и ветра"""
        g = 9.81  # Ускорение свободного падения
        T = control_input  # Подъемная сила

        # Нелинейное уравнение движения по вертикали
        acceleration = (T - self.mass * g) / self.mass  # Вертикальное ускорение
        self.z_dot += acceleration * dt  # Обновление вертикальной скорости

        # Учет ветра (например, как боковая сила)
        wind_effect = self.wind_speed / self.mass  # Влияние ветра на движение
        self.z_dot += wind_effect * dt  # Коррекция вертикальной скорости

        self.z += self.z_dot * dt  # Обновление высоты

        # Проверка на отрицательную высоту
        if self.z < 0:
            self.z = 0
            self.z_dot = 0  # Остановить движение вниз при достижении земли
    def simulate(self, target_z, total_time):
        """Симуляция движения квадрокоптера"""
        dt = 0.01  # Шаг времени
        time_steps = int(total_time / dt)
        z_history = []

        for _ in range(time_steps):
            self.update_wind()  # Обновление ветра
            control_input = self.compute_control_input(target_z)
            self.update_state(control_input, dt)
            z_history.append(self.z)
#            print(f"Высота: {self.z:.2f}, Вертикальная скорость: {self.z_dot:.2f}, Управление: {control_input:.2f}, Ветер: {self.wind_speed:.2f}")  # Логирование данных

        return z_history
# Пример использования нелинейной модели
if __name__ == "__main__":
    quadcopter = QuadcopterNonlinear(mass=1.0, inertia=0.1)  # Масса 1 кг, момент инерции 0.1
    target_height = 5.0  # Желаемая высота
    height_history = quadcopter.simulate(target_z=target_height, total_time=10)  # Симуляция в течение 10 секунд

    # Визуализация результатов
    plt.plot(height_history, label='Высота (z)')
    plt.axhline(y=target_height, color='r', linestyle='--', label='Целевая высота')
    plt.title("Симуляция нелинейной модели квадрокоптера с учетом ветра")
    plt.xlabel("Шаг времени")
    plt.ylabel("Высота (м)")
    plt.legend()
    plt.show()
````

# Линейная модель квадроторора
Поскольку квадротор рассматривается как жесткое тело, динамическое уравнение, описывающее квадротор, можно получить с помощью уравнений Ньютона-Эйлера:
$$
m\ddot{\xi} = G + RT_{B}
$$
где:
- $m$ — масса квадроторора,
- $\ddot{\xi}$ — ускорение квадроторора,
- $G$ — гравитационная постоянная,
- $R$ — матрица вращения,
- $T_{B}$ — подъемная сила роторов.
Для управления квадротором использовался простой линейный контроллер, известный как PID-контроллер, который можно описать следующим образом:
$$
u = -K_pe(t) - K_d \dot{e}(t) - K_i \int_0^t{e(\tau)}d\tau
$$
где:
- $u$ — управляющее воздействие,
- $K_p, K_i, K_d$ — коэффициенты для пропорциональной, интегральной и дифференциальной частей соответственно,
- $e$ — ошибка системы,
- $\tau$ — переменная интегрирования, которая учитывает значения от момента времени \$0$ до текущего момента времени $t$.
### Текущая начальная конфигурация
Квадротор по умолчанию начинает с углов $\phi, \theta, \psi$ и высоты в определенном положении для тестирования:
$$
\begin{align}

  \theta &= -0.9\text{ rad} \\
  \psi &= -3 \text{ rad} \\
  \text{altitude} &= 0\text{ m}
\end{align}
$$
где:
- $\phi$ — угол крена или вращение вокруг оси $x$,
- $\theta$ — угол тангажа или вращение вокруг оси $y$,
- $\psi$ — угол рыскания или вращение вокруг оси $z$.

# Нелинейная модель квадроротора и управления (Quadrotor Model and Control)
## Лагранжиан
Нелинейная модель квадрокоптера может быть выведена с использованием уравнений Эйлера-Лагранжа. Лагранжиан $L$ определяется как сумма кинетической $T$ и потенциальной $U$ энергий:
$$
L = T_T + T_R + U
$$
где:
- $ T_T $ — трансляционная кинетическая энергия,
- $ T_R $ — вращательная кинетическая энергия.
В общем виде лагранжиан может быть записан следующим образом:
$$
L = m \cdot \frac{\dot{x}^2 + \dot{y}^2 + \dot{z}^2}{2} + \frac{1}{2} J \dot{\eta}^2 - mgz
$$
где:
- $ m $ — масса квадрокоптера,
- $ \dot{x}, \dot{y}, \dot{z} $ — скорости по осям $x$, $y$, $z$ соответственно,
- $ J $ — момент инерции,
- $ \eta $ — угловые координаты,
- $ g $ — ускорение свободного падения.
## Уравнение Эйлера-Лагранжа
Уравнение Эйлера-Лагранжа может быть записано как:
$$
\frac{d}{dt} \left( \frac{\partial L}{\partial \dot{\xi}} \right) - \frac{\partial L}{\partial \xi} = F
$$
где $ F $ — это силы и моменты, действующие на тело квадрокоптера.
Силы, действующие на квадрокоптер, могут быть записаны как:
$$
u = f_1 + f_2 + f_3 + f_4
$$
где $ f_i $ — это сила, создаваемая каждым из роторов:
$$
f_i = k \cdot \omega_i^2
$$
Тем временем, моменты по тангажу, рысканию и крену записываются следующим образом:
$$
\tau = \begin{bmatrix}
\tau_\psi \\
\tau_\theta \\
\tau_\phi
\end{bmatrix} = \begin{bmatrix}
(f_2 - f_4) l \\
(f_1 - f_3) l \\
\tau_{M_i}
\end{bmatrix}
$$
где:
- $ l $ — расстояние между роторами и центром масс,
- $ \tau_{M_i} $ — момент, создаваемый двигателем $ i $.
## Уравнения движения
Для позиции квадрокоптера уравнения могут быть записаны как:
$$
\frac{d}{dt} \left( \frac{\partial L_T}{\partial \dot{\xi}} \right) = f
$$
После подстановки получаем:
$$
\frac{d}{dt} \left( m \left( \dot{x}^2 + \dot{y}^2 + \dot{z}^2 \right) + mgz \right) = f
$$
Вычисления приводят к:
$$
m(2\dot{x} + 2\dot{y} + 2\dot{z}) = f + mg
$$
В векторной форме это уравнение выражается как:
$$
\begin{bmatrix}
m \ddot{x} \\
m \ddot{z} - mg
\end{bmatrix}
$$
## Угловые координаты
Для угловых координат уравнения записываются как:
$$
\frac{d}{dt} \left( \frac{\partial L_R}{\partial \dot{R}} \right) = \tau
$$
Подставляя, получаем:
$$
\frac{d}{dt} \left( \frac{1}{2} \dot{\eta}^T J \dot{\eta} \right) = \tau
$$
И вычисляя производные, приходим к следующему уравнению:
$$
J \ddot{\eta} + \dot{J} \dot{\eta} = \tau
$$
## Завершение
Нелинейная динамическая модель квадрокоптера может быть записана в общем виде как:
$$
\begin{bmatrix}
m \ddot{x} \\
m \ddot{z} - mg
\end{bmatrix} = M(η) \ddot{η} + C(η, \dot{η}) \dot{η}
$$
где:
- $ C(η, \dot{η}) = \dot{J} \dot{η} - \frac{1}{2} \frac{\partial}{\partial η} (\dot{η}^T J \dot{η}) $
- $ M(η) = J(η) $
После выполнения некоторых преобразований результирующая система содержит 12 состояний:
$$
X = \begin{bmatrix}
x \\
y \\
z \\
\dot{x} \\
\dot{y} \\
\dot{z} \\
\psi \\
\theta \\
\phi \\
\dot{\psi} \\
\dot{\theta} \\
\dot{\phi}
\end{bmatrix}
$$
и управляемые параметры:
$$
U = \begin{bmatrix}
u_1 \\
u_2 \\
u_3 \\
u_4
\end{bmatrix}
$$


````python
# Реализация нелинейной модели: расчет динамики модели. Для простоты модель будет использовать постоянные значения для массы, момента инерции и других параметров.
#
"""
- **Параметры квадрокоптера**: Определены масса \( m \), ускорение свободного падения \( g \), расстояние до роторов \( l \) и момент инерции \( J \).
- **Начальные условия и управляющие сигналы**: Начальные состояния квадрокоптера устанавливаются в ноль, управляющие сигналы инициализируются значениями для моделирования полета.
- **Обновление состояний**: На каждом шаге времени следующие координаты и скорости обновляются на основе уравнений движения.
- **Графики**: После завершения моделирования результаты выводятся на графиках, показывающих изменения положения и ориентации квадрокоптера с течением времени.

Иллюстрируется  базовая динамика нелинейной модели квадрокоптера, может быть расширена для включения более сложных элементов управления.
"""

import numpy as np
import matplotlib.pyplot as plt

# Параметры квадрокоптера
m = 1.0  # масса
g = 9.81  # ускорение свободного падения
l = 0.2  # расстояние от центра масс до роторов
J = np.diag([0.1, 0.1, 0.1])  # момент инерции

# Время симуляции
dt = 0.01  # шаг времени
t = np.arange(0, 10, dt)  # временной вектор

#print("t =",t)

# Начальные условия
X = np.zeros((12, len(t)))  # [x, y, z, vx, vy, vz, ψ, θ, φ, ωψ, ωθ, ωφ]
U = np.zeros((4, len(t)))  # [u1, u2, u3, u4]

# Установка некоторых управляющих сигналов
U[0, :] = 1.0  # Подъемная сила
U[1, :] = 0.0  # Управление по рысканию
U[2, :] = 0.0  # Управление по тангажу
U[3, :] = 0.0  # Управление по крену

for i in range(1, len(t)):
    # Извлечение текущих значений состояния
    x, y, z, vx, vy, vz, psi, theta, phi, omega_psi, omega_theta, omega_phi = X[:, i - 1]

    # Уравнения движения
    ax = -U[0, i - 1] * np.sin(theta)
    ay = U[0, i - 1] * np.cos(theta) * np.sin(phi)
    az = U[0, i - 1] * np.cos(theta) * np.cos(phi) - m * g

    # Угловое ускорение
    omega_dot = np.linalg.inv(J) @ (np.array([
        (U[1, i - 1] - U[3, i - 1]) * l,
        (U[0, i - 1] - U[2, i - 1]) * l,
        0
    ]))  # Моменты

    # Обновление состояний
    vx += ax * dt
    vy += ay * dt
    vz += az * dt

    x += vx * dt
    y += vy * dt
    z += vz * dt

    omega_psi += omega_dot[0] * dt
    omega_theta += omega_dot[1] * dt
    omega_phi += omega_dot[2] * dt

    psi += omega_psi * dt
    theta += omega_theta * dt
    phi += omega_phi * dt

    # Обновляем вектор состояний
    X[:, i] = np.array([x, y, z, vx, vy, vz, psi, theta, phi, omega_psi, omega_theta, omega_phi])

# Построение графиков результатов
plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.plot(t, X[0, :], label='x (м)')
plt.plot(t, X[1, :], label='y (м)')
plt.plot(t, X[2, :], label='z (м)')
plt.xlabel('Время (с)')
plt.ylabel('Позиция (м)')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(t, X[6, :], label='ψ (рад)', color='r')
plt.plot(t, X[7, :], label='θ (рад)', color='g')
plt.plot(t, X[8, :], label='φ (рад)', color='b')
plt.xlabel('Время (с)')
plt.ylabel('Угол ориентации (рад)')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(t, U[0, :], label='u1')
plt.plot(t, U[1, :], label='u2')
plt.plot(t, U[2, :], label='u3')
plt.plot(t, U[3, :], label='u4')
plt.xlabel('Время (с)')
plt.ylabel('Управляющие сигналы')
plt.legend()

plt.tight_layout()
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

class Quadcopter:
    def __init__(self, mass):
        self.mass = mass

        # Начальное состояние
        self.phi = -0.1
        self.theta = -0.9
        self.psi = -3
        self.altitude = 0

        # Цели
        self.target_phi = -0.3
        self.target_theta = 0
        self.target_psi = -np.pi
        self.target_altitude = 5

        # PID контроллеры
        self.Kp = np.array([1.0, 1.0, 1.0, 1.0])
        self.Ki = np.array([0.0, 0.0, 0.0, 0.0])
        self.Kd = np.array([0.5, 0.5, 0.5, 0.5])

        # Временные параметры
        self.dt = 0.1   # шаг по времени
        self.error_integral = np.zeros(4)
        self.error_previous = np.zeros(4)

    def compute_control_input(self):
        """Вычесление входного контроля на основе PID-контроллера"""
        current_state = np.array([self.phi, self.theta, self.psi, self.altitude])
        target_state = np.array([self.target_phi, self.target_theta, self.target_psi, self.target_altitude])

        # Ошибка
        error = target_state - current_state

        # Пропорциональная часть
        proportional = self.Kp * error

        # Интегральная часть
        self.error_integral += error * self.dt
        integral = self.Ki * self.error_integral

        # Дифференциальная часть
        derivative = self.Kd * (error - self.error_previous) / self.dt

        # Управляющее воздействие
        u = proportional + integral + derivative

        # Обновление предыдущей ошибки
        self.error_previous = error

        return u

    def update_state(self, control_input):
        """Обновление состояния квадрокоптера"""
        # Простая модель динамики
        thrust = control_input[0]  # Входное управление для силы
        self.altitude += thrust * self.dt

        self.phi += 0.1 * control_input[1] * self.dt  # Простая динамика для крена
        self.theta += 0.1 * control_input[2] * self.dt  # Простая динамика для тангажа
        self.psi += 0.1 * control_input[3] * self.dt  # Простая динамика для рыскания

    def simulate(self, total_time):
        """Симуляция полета квадрокоптера"""
        num_steps = int(total_time / self.dt)
        states = []

        for _ in range(num_steps):
            control_input = self.compute_control_input()
            self.update_state(control_input)
            states.append([self.phi, self.theta, self.psi, self.altitude])

        return np.array(states)

# Пример использования
if __name__ == "__main__":
    quadcopter = Quadcopter(mass=1.0)  # Создание квадрокоптера с массой 1 кг
    states = quadcopter.simulate(total_time=10)  # Симуляция полета за 10 секунд

    # Визуализация результатов
    plt.figure(figsize=(12, 8))
    plt.subplot(4, 1, 1)
    plt.plot(states[:, 0], label='Крен (φ)')
    plt.subplot(4, 1, 2)
    plt.plot(states[:, 1], label='Тангаж (θ)')
    plt.subplot(4, 1, 3)
    plt.plot(states[:, 2], label='Рыскание (ψ)')
    plt.subplot(4, 1, 4)
    plt.plot(states[:, 3], label='Высота')
    plt.xlabel('Шаг времени')
    plt.legend()
    plt.tight_layout()
    plt.show()
````

### 1. Огибание препятствий
Огибание препятствий — это маневр, который позволяет БПЛА избегать столкновений с различными объектами, такими как деревья, здания, линии электропередач и другие преграды. Этот маневр может включать следующие аспекты:

- **Обнаружение препятствий**: Использование датчиков, таких как ультразвуковые, инфракрасные или камеры, для обнаружения объектов на пути БПЛА.
- **Планирование маршрута**: Алгоритмы, которые помогают БПЛА пересчитывать маршрут в реальном времени, чтобы избежать обнаруженных препятствий.
- **Управление полетом**: Изменение высоты, направления и скорости полета для успешного обхода препятствий. Это может включать маневры, такие как повороты, подъемы или спуски.

### 2. Пикирование
Пикирование — это маневр, при котором БПЛА резко снижает высоту, часто с целью быстрого приближения к земле или для выполнения задачи, такой как фотографирование или сброс полезной нагрузки. Этот маневр включает:

- **Резкое снижение высоты**: БПЛА изменяет угол атаки и увеличивает скорость для быстрого снижения.
- **Контроль устойчивости**: Важно поддерживать управление и стабильность БПЛА во время пикирования, чтобы избежать потери контроля.
- **Использование датчиков**: Датчики высоты и скорости помогают пилоту или автоматической системе контролировать параметры полета во время пикирования.

Оба маневра требуют высокой степени автоматизации и точности в управлении, особенно в сложных условиях, что делает их важными аспектами проектирования и программирования БПЛА.

 #Список препятствий
 представлялся в виде массива (списка) с заранее заданными координатами, где каждое препятствие будет содержать информацию о своей позиции и расстоянии до него. Дрон будет проверять наличие препятствия в заданной позиции перед тем, как двигаться вперед.

##Препятствия должны иметь координаты x и y, а также height, чтобы дрон мог их правильно "видеть" и реагировать на них.

##Реакция дрона на препятствия: Дрон двигается вперед, если расстояние до препятствия больше допустимого, и подниматься, если расстояние меньше определенного значения.

Логика движения:

Дрон должен проверять расстояние до препятствий вблизи своей текущей позиции.

Если дрон поднимается, его высота должна увеличиваться, а координаты по оси x увеличиваться только при движении вперед.


````python
import random
import time
import matplotlib.pyplot as plt

class Drone:
    def __init__(self, obstacles):
        self.position = (0, 0)  # Начальная позиция дрона (x, y)
        self.height = 0  # Начальная высота дрона
        self.speed = 1  # Скорость дрона
        self.path = [(0, 0, 0)]  # Путь дрона (x, y, height)
        self.obstacles = obstacles  # Список препятствий

    def get_lidar_distance(self):
        # Проверяем дистанцию до ближайшего препятствия
        min_distance = float('inf')  # Начальное значение бесконечность
        for obs in self.obstacles:
            dx = self.position[0] - obs['x']
            dy = self.position[1] - obs['y']
            distance = (dx**2 + dy**2)**0.5

            # Если высота дрона меньше высоты препятствия, учитываем расстояние
            if self.height < obs['height'] and distance < min_distance:
                min_distance = distance

        return min_distance  # Возвращаем расстояние до ближайшего препятствия

    def move_forward(self):
        x, y = self.position
        self.position = (x + self.speed, y)
        # Добавление текущей высоты к пути
        self.path.append((self.position[0], self.position[1], self.height))
        print(f"Дрон перемещается вперед. Новая позиция: {self.position}, высота: {self.height}")

    def turn_left(self):
        print("Дрон поворачивает влево.")

    def climb_height(self):
        self.height += 1  # Увеличиваем высоту
        print(f"Дрон поднимается. Новая высота: {self.height}")

    def avoid_obstacle(self):
        distance = self.get_lidar_distance()
        print(f"Расстояние до ближайшего препятствия: {distance:.2f} см")

        if distance < 3:  # Если расстояние до ближайшего препятствия меньше 3
            self.turn_left()  # Поворачиваем влево (или в другую сторону)
            self.climb_height()  # Поднимаемся, чтобы избежать препятствия
        else:
            self.move_forward()  # Двигаемся вперед

def plot_drone_path(path, obstacles):
    x_vals, _, heights = zip(*path)

    plt.figure(figsize=(10, 6))
    plt.plot(x_vals, heights, marker='o', label="Путь дрона по высоте", color='blue')

    obs_x = [obs['x'] for obs in obstacles]
    obs_y = [obs['y'] for obs in obstacles]
    obs_height = [obs['height'] for obs in obstacles]
    plt.scatter(obs_x, obs_height, marker='x', color='red', s=100, label='Препятствия')

    plt.xlim(0, max(x_vals) + 1)
    plt.ylim(0, max(heights) + 1)
    plt.title("Путь дрона и препятствия")
    plt.xlabel("X (положение)")
    plt.ylabel("Высота")
    plt.axhline(0, color='gray', lw=0.5, ls='--')
    plt.legend()
    plt.grid()
    plt.show()

def create_obstacles(num_obstacles, interval):
    obstacles = []
    for i in range(num_obstacles):
        obs_x = i * interval  # Располагаем препятствия через заданный интервал
        obs_height = random.randint(1, 3)  # Высота препятствий от 1 до 3
        obstacles.append({'x': obs_x, 'y': 0, 'height': obs_height})
    return obstacles

def main():
    num_obstacles = 5  # Количество препятствий
    interval = 4  # Интервал между препятствиями

    obstacles = create_obstacles(num_obstacles, interval)

    drone = Drone(obstacles)

    for _ in range(20):  # Цикл управления дроном
        drone.avoid_obstacle()  # Проверка на препятствие
        time.sleep(0.5)  # Пауза для следующего обновления

    plot_drone_path(drone.path, drone.obstacles)  # Визуализация пути дрона и препятствий

if __name__ == "__main__":
    main()
````

Для реализации динамики БПЛА с целью определения оптимального времени и высоты для выполнения манёвра пикирования, необходимо учитывать многие факторы, такие как скорость, высота, угол атаки и время снижения. В этом примере мы создадим простую модель, в рамках которой БПЛА будет снижаться с заданной высоты к земле с возможностью контролировать угол пикирования и скорость.

Принципы моделирования
Состояние БПЛА: Определяется высотой, скоростью и временем.

Динамика снижения: Определим уравнения движения БПЛА, которые будут учитывать силы, действующие на него во время пикирования.

Оптимизация времени и высоты: Установим цели для снижения, которые позволят БПЛА выполнить пикирование эффективно.

##Угол тангажа (Pitch Angle):
Угол тангажа: Это угол между осью самолета (обычно между продольной осью) и горизонтом. Угол тангажа показывает, насколько нос самолета поднят или опущен относительно горизонтальной линии. Увеличение угла тангажа означает, что нос самолета поднимается вверх, а уменьшение — что он опускается вниз. Угол тангажа влияет на подъемную силу и, соответственно, на вертикальное движение самолета.

##Угол пикирования (Dive Angle) или угол атаки:
Угол пикирования: Это угол между продольной осью самолета и направлением его движения (траекторией полета). Угол пикирования показывает, насколько самолет наклонен вниз относительно направления своего движения. Пикирование происходит, когда самолет наклоняется вниз, и это может привести к снижению высоты. Угол пикирования важен для маневров, таких как снижение или атака.

##Таким образом, основное различие между этими углами заключается в том, что угол тангажа измеряет наклон самолета относительно горизонта, а угол пикирования — наклон самолета относительно его траектории полета.


При моделировании динамики БПЛА важно учитывать оба этих угла, поскольку они влияют на поведение аппарата в воздухе. Вы можете реализовать управление этими углами через параметры вашего симуляционного кода, чтобы получить более реалистичное поведение во время пикирования и других маневров.


````python
import numpy as np
import matplotlib.pyplot as plt

class UAV:
    def __init__(self, initial_height, initial_speed, descent_rate, target_height):
        self.height = initial_height  # Высота БПЛА
        self.speed = initial_speed  # Начальная скорость
        self.descent_rate = descent_rate  # Скорость снижения
        self.target_height = target_height  # Целевая высота
        self.time = 0  # Время
        self.path = []  # Для хранения пути (время, высота)

    def update(self, dt):
        # Обновление состояния БПЛА
        self.height -= self.descent_rate * dt  # Обновляем высоту
        self.time += dt  # Увеличиваем время
        self.path.append((self.time, max(self.height, 0)))  # Сохраняем данные

    def is_target_reached(self):
        return self.height <= self.target_height

def simulate_UAV(initial_height, initial_speed, descent_rate, target_height, total_time):
    dt = 0.1  # Временной шаг
    uav = UAV(initial_height, initial_speed, descent_rate, target_height)

    while uav.height > 0 and uav.time < total_time:
        uav.update(dt)

    return uav.path

def plot_path(path):
    times, heights = zip(*path)  # Разделяем время и высоту
    plt.plot(times, heights, label="Путь БПЛА", color='blue')
    plt.axhline(y=0, color='gray', linestyle='--', label="Земля")
    plt.axhline(y=target_height, color='red', linestyle='--', label="Целевая высота")
    plt.xlabel("Время (с)")
    plt.ylabel("Высота (м)")
    plt.title("Динамика снижения БПЛА")
    plt.legend()
    plt.grid()
    plt.show()

# Параметры симуляции
initial_height = 1000  # Начальная высота (м)
initial_speed = 50  # Начальная скорость (м/с)
descent_rate = 5  # Скорость снижения (м/с)
target_height = 100  # Целевая высота для пикирования (м)
total_time = 300  # Общее время симуляции (с)

# Симуляция
path = simulate_UAV(initial_height, initial_speed, descent_rate, target_height, total_time)

# Построение графика
plot_path(path)
````
